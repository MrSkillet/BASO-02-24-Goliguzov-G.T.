```python
import numpy as np # Импортируем numpy
a = np.zeros((6, 6)) # Создаём массив из 0
b = np.ones(np.shape(a)) # Создаём массив из 1 с той же размерностью
c = np.arange(-10, 10, 5) # Создаём одномерный массив от -10 до 10 с шагом 5
print(a, b, c, sep='\n') # Выводим массивы
print(np.shape(a), np.dtype(a[0][0]), sep='\n') # Выводим размерность и тип первой переменной для первого массива
a[0][0], a[1][0],a[2][0],a[3][0],a[4][0],a[5][0] = 1, 2, 3, 4, 5, 6 # Изменяем первый элемент каждой строки на её номер
print(a[1:5:2]) # Делаем срез с 1 по 5 стороку с шагом в 2
print((a + 3) * 2) # Выводим поэлементную сумму первого массива и 3, умноженную на 2
print(np.mean(a)) # Выводим среднюю величину массива
```
