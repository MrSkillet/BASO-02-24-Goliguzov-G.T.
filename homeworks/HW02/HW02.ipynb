```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/mirea-aie-2025/aie-course-meta/refs/heads/main/seminars/S02/S02-hw-dataset.csv'
df = pd.read_csv(url)
print(df.head())
print('=' * 52)
print(df.info())
print('=' * 52)
print(df.describe())
print('=' * 52)
print(df.isna().mean())
print('=' * 52)
print(df.duplicated())
print('=' * 52)
print('Аномальный возраст:')
print(df[df.age > 100])
print(df[df.age != df.age])
print('=' * 52)
print('Аномальные покупки:')
print(df[df.purchases < 0])
print('=' * 52)
print('Аномальный доход:')
print(df[df.revenue < df.purchases])
# Были обнаружены неполные (NaN) и нереалистичные (120) данные о возрасте, невозможное колличество покупок (-1) и аномальные данные о зароботке (0 при наличаи покупок).
print('=' * 52)
print(df.value_counts(df.country))
print('=' * 52)
print(df.groupby(['country'])['revenue'].mean())
# Было расчитано число пользователей по странам и среднее значение дохода по каждой из них. Самая 'популярная' страна - Россия (RU), самый большой средний доход в Дании (DE).
plt.hist(df.country, label="Распределение по странам", color='r')
plt.xlabel("Страна")
plt.ylabel("Количество пользователей")
plt.legend()
plt.savefig('f1')
plt.show()
# Строим график распределения пользователей по мтранам.
plt.boxplot(df['revenue'])
plt.xlabel("Группа")
plt.ylabel("Доход")
plt.savefig('f2')
plt.show()
# Строим боксплот по доходу.
plt.scatter(df.purchases, df.revenue, label="Зависимость дохода от числа покупок", color='r')
plt.ylabel("Доход")
plt.xlabel("Покупки")
plt.legend()
plt.savefig('f3')
plt.show()
# Строим точечный график зависимости дохода от числа покупок.
```
